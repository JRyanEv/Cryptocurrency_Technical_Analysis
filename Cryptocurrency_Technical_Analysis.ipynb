```python
from tradingview_ta import TA_Handler, Interval, Exchange
import time

def crypto_KUCOIN_1_DAY(symbol_IN):
    
    print(f"\nKUCOIN ASSET: {symbol_IN.upper()} 1 DAY BAR\n")
    
    KUCOIN_ASSET = TA_Handler(
    symbol="{x}".format(x = symbol_IN + "USDT"),
    screener="crypto",
    exchange="KUCOIN",
    interval=Interval.INTERVAL_1_DAY
    )

    print("Opening Price")
    print(KUCOIN_ASSET.get_analysis().indicators["open"])
    print("\n")

    print("Closing Price")
    print(KUCOIN_ASSET.get_analysis().indicators["close"])
    print("\n")

    print("Technicals")
    print(KUCOIN_ASSET.get_analysis().summary)
    print("\n")

    print("RSI")
    print(KUCOIN_ASSET.get_analysis().indicators["RSI"])
    print("\n")

    print("MACD")
    print(KUCOIN_ASSET.get_analysis().indicators["MACD.macd"])
    
def crypto_KUCOIN_1_HOUR(symbol_IN):
    
    print(f"\nKUCOIN ASSET: {symbol_IN.upper()} 1 HOUR BAR\n")
    
    KUCOIN_ASSET = TA_Handler(
    symbol="{x}".format(x = symbol_IN + "USDT"),
    screener="crypto",
    exchange="KUCOIN",
    interval=Interval.INTERVAL_1_HOUR
    )

    print("Opening Price")
    print(KUCOIN_ASSET.get_analysis().indicators["open"])
    print("\n")

    print("Closing Price")
    print(KUCOIN_ASSET.get_analysis().indicators["close"])
    print("\n")

    print("Technicals")
    print(KUCOIN_ASSET.get_analysis().summary)
    print("\n")

    print("RSI")
    print(KUCOIN_ASSET.get_analysis().indicators["RSI"])
    print("\n")

    print("MACD")
    print(KUCOIN_ASSET.get_analysis().indicators["MACD.macd"])
    
#User Enters Symbol such as 'btc', 'eth', 'sol'
user_symbol = input("Enter Symbol: ")
crypto_KUCOIN_1_DAY(user_symbol)
crypto_KUCOIN_1_HOUR(user_symbol)
```
